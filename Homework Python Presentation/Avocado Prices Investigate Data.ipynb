```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 9))

for key in subsetAvos1_df['Geography'].unique():
    print(key)

    axes.plot(subsetAvos1_df[subsetAvos1_df['Geography'] == key]['Current Year Week Ending'],
              subsetAvos1_df[subsetAvos1_df['Geography'] == key]['ASP Current Year'],
              marker='o',
              linestyle='-',
              markersize=5,
              label=key)
    
plt.legend()
```

```python
grouped = subsetAvos1_df.groupby('Geography')
```

```python
plt.figure(figsize=(20, 5)) 

for group_name, group_data in grouped:
    plt.plot(group_data['Current Year Week Ending'], group_data['ASP Current Year'], label=group_name)

plt.title('Avocado Price Trends by Geography')
plt.xlabel('Date')
plt.ylabel('Average Selling Price')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Add a legend
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

plt.tight_layout()
plt.show()
```

```python
# yuk. I think I need to group the data by month and take an average for each month? 
```

```python
subsetAvos1_df['month_year'] = subsetAvos1_df['Current Year Week Ending'].dt.to_period('M')  # Extract month and year
grouped = subsetAvos1_df.groupby(['Geography', 'month_year'])
```

```python
average_prices = grouped['ASP Current Year'].mean().reset_index()
```

```python
subsetAvos1_df.dtypes
```

```python
plt.figure(figsize=(12, 6))

for group_name, group_data in average_prices.groupby('Geography'):
    plt.plot(group_data['month_year'], group_data['ASP Current Year'], label=group_name)

plt.title('Average Avocado Prices by Geography (Monthly)')
plt.xlabel('Month')
plt.ylabel('Average Selling Price')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Convert to string because of 'period' error
subsetAvos1_df['monthYear'] = subsetAvos1_df['Current Year Week Ending'].dt.to_period('M').dt.strftime('%Y-%m')  

grouped = subsetAvos1_df.groupby(['Geography', 'monthYear'])
average_prices = grouped['ASP Current Year'].mean().reset_index()

plt.figure(figsize=(12, 6))

for group_name, group_data in average_prices.groupby('Geography'):
    plt.plot(group_data['monthYear'], group_data['ASP Current Year'], label=group_name)

plt.title('Average Monthly Avocado Prices by US Geographical Areas of Sale YTD 2023')
plt.xlabel('Month')
plt.ylabel('Average Selling Price USD')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
